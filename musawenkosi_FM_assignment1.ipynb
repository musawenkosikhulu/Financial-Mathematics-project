```python
import numpy as np  #import numpy library for using maths operations
import matplotlib.pyplot as plt # import matplotlib for visualizations


def strike_function(Number):  #This creates a function plots
    pay_off = np.zeros(100) # This line of code is the payoff assignemnt.
    stock_prices = np.linspace(0,200,100) # This line of code assignment of  stock prices
    
    for i in range(100): #This loop for iterating through the stock prices starting from zero
        if Number >= stock_prices[i]: #This line of code checks if the condition N is greater or equal to stock prices
            pay_off[i] = 0 #If the condition is met we assign zero
            
        else:
            pay_off[i] = stock_prices[i]-Number #If N is less than stock prices we assign it to pay off
    return pay_off #This function results the pay-off as we wanted

option_1 = strike_function(50) #This line call the function in order to plot for option 1
option_2 = strike_function(60) #This line call the function in order to plot for option 2
option_3 = strike_function(70) #This line call the function in order to plot for option 3

stock_prices = np.linspace(0,200,100) #This line of code creates the stock prices steps
plt.title('Payoff vs Stock prices')
plt.ylabel('Payoff')
plt.xlabel('stock prices')
plt.plot(stock_prices,option_1, color = 'red')
plt.plot(stock_prices,option_2, color = 'yellow')
plt.plot(stock_prices,option_3, color = 'black')
            
```

```python

Pay =option_1 -2*option_2 + option_3 #This line of code calculating the pay off

plt.plot(stock_prices ,  Pay, label = "Portfolio",color = "red") #This line of code plots for option 1
plt.xlabel('Stock prices)') #This line of code is for x label
plt.ylabel('Portfolio') #This line of code is for y label
plt.title('Portfolio of the P price movements in the stock market') #This line of code is for title
plt.legend() #This line of code is for the legends
```

```python
def print_statistics(array):  
    ''' Prints selected statistics.
    Parameters
    ==========
    array: ndarray       (object to generate statistics on)
    '''
    sta = scs.describe(array)
    print ("%14s %15s" % ('statistic', 'value'))
    print (30 * "-")
    print ("%14s %15.5f" % ('size', sta[0]))
    print ("%14s %15.5f" % ('min', sta[1][0]))
    print ("%14s %15.5f" % ('max', sta[1][1]))
    print ("%14s %15.5f" % ('mean', sta[2]))
    print ("%14s %15.5f" % ('std', np.sqrt(sta[3])))
    print ("%14s %15.5f" % ('skew', sta[4]))
    print ("%14s %15.5f" % ('kurtosis', sta[5]))
```

```python
symbols=['Adj Close']
for sym in symbols:
    print ("\nResults for symbol %s" % sym)
    print (30 * "-")
    log_data = np.array(log_returns['Adj Close'].dropna())
    print_statistics(log_data)
```

```python
def normality_tests(arr):
    ''' Tests for normality distribution of given data set.
    Parameters
    ==========
    array: ndarray
    object to generate statistics on
    '''
    print ("Skew of data set %14.3f" % scs.skew(arr))
    print ("Skew test p-value %14.3f" % scs.skewtest(arr)[1])
    print ("Kurt of data set %14.3f" % scs.kurtosis(arr))
    print ("Kurt test p-value %14.3f" % scs.kurtosistest(arr)[1])
    print ("Norm test p-value %14.3f" % scs.normaltest(arr)[1])
```

```python
scs.probplot(log_returns['Adj Close'].dropna(), plot=plt,fit=True)
plt.grid(True)
plt.xlabel('theoretical quantiles')
plt.ylabel('sample quantiles')
for sym in symbols:
    print ("\nResults for symbol %s" % sym)
    print (32 * "-")
    log_data = np.array(log_returns[sym].dropna())
    normality_tests(log_data)
```
